```python
#tamanho do plot
plt.figure(figsize=(11,4))
#titulo do plot
plt.title('Quantidade De Vendas Globais (MI)', loc='left')
#plot de barras 
sns.barplot(data=df,x='Year',y='Global', ci=None,color='#5b4', estimator=sum);
#mudança do y label titulo 
plt.ylabel("Quantidade De Vendas(MI)");
```

```python
 df = df.loc[(df['Year']!= 2019)&(df['Year']!= 2020)]
```

```python
df.groupby(by=['Year']).sum()
```

```python
plt.figure(figsize=(10,4))
plt.title('Analise Da Distribuiçao Global')
sns.boxplot(data=df,x='Year',y='Global');
```

```python
df.loc[df['Global']>=10]
```

```python
Analise = df.groupby( by=['Year'] ).sum().reset_index()
```

```python
Analise
```

```python
America = (Analise['North America']/Analise['Global'])*100
Europa = (Analise['Europe']/Analise['Global'])*100
Japao = (Analise['Japan']/Analise['Global'])*100
Mundo = (Analise['Rest of World']/Analise['Global'])*100
America,Europa,Japao,Mundo
```

```python
plt.figure(figsize=(10,4))
rotulos = Analise['Year']
grupos = [0,1,2,3,4,5]
plt.title('analise da distribuiçao por continentes');
#america
plt.bar(grupos,America, color='#b2f4b0',edgecolor='white');
#europa
plt.bar(grupos,Europa,bottom=America,color='#f9bc86' ,edgecolor='white');
#Japao 
plt.bar(grupos,Japao,bottom=[A+B for A,B in zip(America,Europa)],color='#a3acff' ,edgecolor='white');
#Resto do Mundo
plt.bar(grupos,Mundo,bottom=[A+B+C for A,B,C in zip(America,Europa,Japao)],color='#d3acfe' ,edgecolor='white');
#labels 
plt.xticks(grupos,rotulos);
plt.xlabel('Grupo')
plt.ylabel('Distribuiçao % ')
#legenda
plt.legend(['America','Europa','Japao','Mundo']);
```

```python
from sklearn.preprocessing import LabelEncoder
funçao_label = LabelEncoder()

df['Produtor']=funçao_label.fit_transform(df['Publisher'])
df['Genero']=funçao_label.fit_transform(df['Genre'])
df['Jogo']=funçao_label.fit_transform(df['Game'])
df.head()
```

```python
 paleta= sns.color_palette('BuGn',10)
paleta 
```

```python
plt.figure(figsize=(20,6))
plt.title('Analise Dos Produtores(MI)')
sns.scatterplot(data=df,x='Produtor',y='Global', color=paleta[6] );
```

```python
plt.figure( figsize=(20,5))
plt.title('Análise dos generos do Game (mi)')
sns.scatterplot(data=df, x='Genero', y='Global', color=paleta[5] );
```

```python
plt.figure( figsize=(20,5))
plt.title('Análise dos Games (mi)')
sns.scatterplot(data=df, x='Jogo', y='Global', color=paleta[7] );
```
